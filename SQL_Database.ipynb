```python
def query(query_a_ejecutar):
  connection=sqlite3.connect("Diplo_DB.db")
  cursor = connection.cursor()
  try: #Ya que está hago uso de try y catch para obtener una respuesta adecuada ante errores
    cursor.execute(f'''{query_a_ejecutar}''')
    result = cursor.fetchall()
  except sqlite3.OperationalError:
    print("Error al procesar la query")
    result = -1 #En caso de error devolverá el código de error -1
  finally:
    cursor.close()
    connection.close()
    return result
```

```python
def update(query_a_ejecutar):
  connection=sqlite3.connect("Diplo_DB.db")
  cursor = connection.cursor()
  try: #Ya que está hago uso de try y catch para obtener una respuesta adecuada ante errores
    cursor.execute(f'''{query_a_ejecutar}''')
    connection.commit() #Persisto el cambio en la base
    print('Campo actualizado')
    result = 1
  except sqlite3.Error as error:
    print(f'Error al actualizar: {error}')
    result = -1 #En caso de error devolverá el código de error -1
  finally:
    cursor.close()
    connection.close()
    return result
```

2. Se quiere saber cuál es el país del cliente que más dinero gastó en una sola compra.

```python
query = '''SELECT  c.Pais, max(d.Precio) 
        FROM Facturas f
        LEFT JOIN Clientes c
        ON c.ClienteID = f.ClienteID
        LEFT JOIN Detalles d
        ON d.FacturaID = f.FacturaID
        '''

query_df(query)
```

Usando MySQL la función MAX tira error, pero GROUP BY / ORDER BY / LIMIT 1 saca el mismo resultado

```python
query = '''SELECT c.Pais, MAX(d.precio) as max
FROM Clientes c
LEFT JOIN Facturas f
ON c.ClienteID = f.ClienteID
LEFT JOIN Detalles d
ON f.FacturaID = d.FacturaID
GROUP BY c.Pais
ORDER BY max desc
LIMIT 1'''

query_df(query)
```
