# **Stores analysis**
---

The end goal of this analysis is to find data that will help a new store owner make a decisions about the rental space. The questions we were trying to answer were:


*   Is there a correlation between the size of the store and its sales/customers?
*   Is there a correlation between the items available and their sales/customers


 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
stores_data= pd.read_csv("/content/drive/MyDrive/csv/Stores.csv")
stores_data.head(15)
```

```python
stores_data.info()
```

```python
stores_data.duplicated()
```

```python
stores_data.describe()
```

```python
stores_data['Monthly_Customers_Count'] = stores_data['Daily_Customer_Count'] * 30
stores_data['Avg_Customer_Spend_Month'] = stores_data['Store_Sales'] / stores_data['Daily_Customer_Count']
stores_data['Avg_daily_sales'] = stores_data['Store_Sales'] / 30
```

```python
stores_data.head(10)
```

```python
stores_data=stores_data.round(decimals=2)
stores_data.head()
```

```python
fig = plt.figure(figsize=(10, 7))
y = stores_data["Store_Area"]
x = stores_data["Items_Available"]
plt.scatter(x, y, marker=".", color="g")
plt.title("Area and numbers of items correlation", size=16)
plt.ylabel("Stores area", size=12)
plt.xlabel("Items available", size=12)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
y = stores_data["Store_Area"]
x = stores_data["Monthly_Customers_Count"]
plt.scatter(x, y, marker=".", color="g")
plt.title("Area and monthly customers correlation", size=16)
plt.ylabel("Stores area", size=12)
plt.xlabel("Monthly customers", size=12)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
y = stores_data["Items_Available"]
x = stores_data["Monthly_Customers_Count"]
plt.scatter(x, y, marker=".", color="g")
plt.title("Items available and monthly customers correlation", size=16)
plt.ylabel("Items_Available", size=12)
plt.xlabel("Monthly customers", size=12)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
y = stores_data["Items_Available"]
x = stores_data["Store_Sales"]
plt.scatter(x, y, marker=".", color="g")
plt.title("Items available and store sales correlation", size=16)
plt.ylabel("Items_Available", size=12)
plt.xlabel("Store_Sales", size=12)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
y = stores_data["Store_Area"]
x = stores_data["Store_Sales"]
plt.scatter(x, y, marker=".", color="g")
plt.title("Store area and store sales correlation", size=16)
plt.ylabel("Items_Available", size=12)
plt.xlabel("Store_Sales", size=12)
plt.show()
```

```python
df= pd.DataFrame(stores_data)
df.to_csv("./sample_data/stores_data.csv")
```

###Conclusions:


*   There is no correlation between the store area and the store sales/customers.
*   There is no correlation between the number of items available and the store sales/customers.



###Recommendations:

*   As the results show there is no need for a big rental space for the first store.
